```python
# caminhos
source_path = "/Volumes/sptrans_catalog/gtfs_data/landing_zone/"
target_table = "sptrans_catalog.gtfs_data.bronze_stop_times"
checkpoint_path = "/Volumes/sptrans_catalog/gtfs_data/landing_zone/_checkpoints/bronze_stop_times"

# Leitura com Auto Loader
df_bronze = (spark.readStream
  .format("cloudFiles")
  .option("cloudFiles.format", "csv")
  .option("header", "true")  
  .option("pathGlobFilter", "stop_times.txt")
  .option("cloudFiles.schemaLocation", checkpoint_path)
  .load(source_path))
  
# Escrita no Unity Catalog
(df_bronze.writeStream
.option("checkpointLocation", checkpoint_path)
  .outputMode("append")
  .toTable(target_table))
```
